# 06_ecommerce_recommendation_system — Data Exploration

This notebook explores the implicit feedback dataset (purchases) and the product catalog.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

ROOT = Path('..').resolve()
purchases = pd.read_csv(ROOT/'data'/'purchases.csv')
products = pd.read_csv(ROOT/'data'/'products.csv')
purchases.head(), products.head()
```

```python

print('Rows purchases:', len(purchases))
print('Unique users:', purchases['user_id'].nunique())
print('Unique products purchased:', purchases['product_id'].nunique())
```

## Purchases per user / product

```python

per_user = purchases.groupby('user_id')['product_id'].nunique()
per_item = purchases.groupby('product_id')['user_id'].nunique()

plt.figure()
plt.hist(per_user, bins=20)
plt.title('Distinct products purchased per user')
plt.xlabel('Count'); plt.ylabel('Users')
plt.show()

plt.figure()
plt.hist(per_item, bins=20)
plt.title('Users per product')
plt.xlabel('Count'); plt.ylabel('Products')
plt.show()
```
