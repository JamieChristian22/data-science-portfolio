## Quick relationship checks

```python

# Automatically plot up to 6 numeric columns vs target (boxplots)
import math
num_cols = [c for c in df.columns if df[c].dtype != 'object' and c not in ['customer_id','application_id','transaction_id']]
cols = [c for c in num_cols if c != target][:6]
for c in cols:
    plt.figure()
    df.boxplot(column=c, by=target)
    plt.title(f'{c} by {target}')
    plt.suptitle('')
    plt.xlabel(target)
    plt.ylabel(c)
    plt.show()
```
